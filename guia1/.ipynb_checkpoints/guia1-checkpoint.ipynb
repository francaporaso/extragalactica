```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.cosmology import FlatLambdaCDM
```

```python
cosmo = FlatLambdaCDM(H0=70.0, Om0=0.3)
```

```python
fortran_file = pd.read_csv('cosmo.dat', header=None, delimiter=',', names=['z', 'chi', 'd_L', 'd_A'])
```

```python
z = np.arange(0.0,6.0,0.001)
```

```python
plt.plot(z, cosmo.luminosity_distance(z), label='d_L py')
plt.plot(fortran_file.z, fortran_file.d_L, label='d_L for')
plt.plot(z, cosmo.angular_diameter_distance(z), label='d_A py')
plt.plot(fortran_file.z, fortran_file.d_A, label='d_A for')
plt.ylim(0.0,4000)
plt.xlim(0.0,3)
plt.legend()
```

```python
plt.plot(z, (cosmo.luminosity_distance(z).value - fortran_file.d_L)/cosmo.luminosity_distance(z).value, '.', label='d_L diff')
#plt.plot(fortran_file.z, , label='d_L for')
#plt.plot(z, (cosmo.angular_diameter_distance(z).value - fortran_file.d_A)/cosmo.angular_diameter_distance(z).value, '.', label='d_A diff')
#plt.plot(fortran_file.z, , label='d_A for')
#plt.ylim(0.0,4000)
#plt.xlim(0.0,3)
plt.legend()
plt.semilogy()
```
